```python
# Setting up the imports 

from steemdata import SteemData
import datetime
from datetime import datetime as dt

import numpy as np

import pandas as pd
```

```python
# Init connection to database

db = SteemData()
```

```python
query = {
    'type' : 'vote',
    'timestamp' : {'$gte': dt.now() - datetime.timedelta(days=1)},
    'author': 'alexeykh'}
    
proj = {'_id': 0, 'timestamp': 1, 'voter' : 1, 'weight': 1}

sort = [('timestamp', -1)]
```

```python
%%time

result = db.Operations.find(query,
                            projection=proj,
                            sort=sort)

res = list(result)
```

```python
Data = pd.DataFrame({'voter': [e['voter'] for e in res],
                     'weight': [e['weight']/100 for e in res]},
              index=[e['timestamp'] for e in res])
```

```python
len(Data)
```

```python
VoteSort = Data.groupby('voter').sum().sort_values('weight', ascending=False)

VoteSort
```
